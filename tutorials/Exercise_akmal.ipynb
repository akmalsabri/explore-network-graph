### Data Manipulation

## Combining the data 
- Combine the Export and import dataframe

```python
df_ex_im = pd.concat([df_import, df_export["Total Export (MYR)"]], axis=1)
df_ex_im
```

### Additional Data
- Balance of Trade(BOT) is Value of Import - Value of Export
- Create a new column named BOT 
- View the row with the maximum BOT
- View the row with the minimum number of BOT

```python
# Create a new column named BOT
df_ex_im["BOT"] = df_import["Total Import (MYR)"] - df_export["Total Export (MYR)"]
df_ex_im.head()
```

```python
df_ex_im.dtypes
```

```python
# View the row with the maximum BOT

df_ex_im[df_ex_im["BOT"] == max(df_ex_im["BOT"])]
```

```python
# View the row with the minimum number of BOT

df_ex_im[df_ex_im["BOT"] == min(df_ex_im["BOT"])]
```

### Groupby
