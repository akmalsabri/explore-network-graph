```python
from pathlib import Path
import os
from urllib.error import HTTPError
import urllib
import urllib.request
import bs4 as bs4
import time
import pandas as pd 
import time

# library for sentiment analysis

from pytorch_pretrained_bert.modeling import BertForSequenceClassification
from finbert.finbert import predict

model = BertForSequenceClassification.from_pretrained('models/classifier_model/finbert-sentiment', num_labels=3, cache_dir=None)
```

```python
# upload csv 

df_title = pd.read_csv('isaham_berita.csv')
```

```python
df_title.head(10)
```

```python
# convert title to list


list_tajuk = df_title['Title'].tolist()
```

```python
def main():


    list_pred = []
    list_conf = []

    for tajuk in list_tajuk :

        print(tajuk)

        sentence , prediction , confidence = predict(tajuk,model)

        prediction = str(prediction)
        prediction = prediction.replace("[","").replace("]","").replace("'","")
        list_pred.append(prediction)

        confidence = str(confidence)
        confidence = confidence.replace("[","").replace("]","").replace("'","")
        list_conf.append(confidence)


    return list_pred, list_conf
```

```python
if __name__ == "__main__":

    tic = time.clock()

    pred , conf = main()

    dict_df = {'Prediction':pred, 'Confidence':conf}

    df_pred = pd.DataFrame(dict_df)

    # cantum dua dataframe

    df_final = pd.concat([df_title,df_pred], axis= 1)

    df_final.to_csv('pred_title_isaham.csv')

    toc = time.clock()

    print(toc-tic)
```
